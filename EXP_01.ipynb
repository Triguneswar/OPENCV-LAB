```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()

cap = cv2.VideoCapture('your_video.mp4')

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    cv2_imshow(frame)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
